```python
import pandas as pd
df_1=pd.read_csv("CPI.csv", encoding = "ISO-8859-1")
df_1.head(100)
```

```python
df_2 = df_1.melt(
    id_vars = ['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'],
    value_vars = df_1.columns[4:], var_name = 'year').sort_values(['Country Name', 'year'])
df_2.head()
```

```python
df_3 = df_2
```

```python
df_4 = df_3[['Country Name','year',"value"]]
df_4.columns
```

```python
df_4 = df_4.set_index("year")
df_4 
```

```python
df_5 = pd.read_csv("GDP_per_Capita.csv")
df_6 = df_4.rename(columns={"value":"CPI"})
year = [1990,1995,2000,2005,2010,2015,2017]
df_7=df_6.reset_index()
df_7.columns
```

```python
df_8 = df_7[pd.to_numeric(df_7["year"]).isin(year)]
df_8['year']=df_8['year'].astype(int)
df_9 = pd.read_csv("GDP_per_Capita.csv")
```

```python
df_9.head()
```

```python
df_9["year"]=df_9["year"].astype(int)
df_10_merge = pd.merge(df_9,df_8,left_on=['country', 'year'],right_on=['Country Name', 'year'],how = "inner")
df_10_merge
```

```python
df_11=df_10_merge.sort_values(['country', 'year']).drop('Country Name', axis=1).set_index("year")
df_11["CPI"].fillna("Data Unavailable",inplace = True)
df_11.head()
```

```python
df_11.to_csv("project_data.csv")
```
