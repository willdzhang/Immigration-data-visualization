```python
import pandas as pd
import numpy as np
import requests
import json
from pprint import pprint
import matplotlib.pyplot as plt
```

```python
un_migrant = pd.read_csv("Resources/UN_MigrantStockByOriginAndDestination_2017.csv", encoding = "ISO-8859-1")

countries_df = pd.read_csv("Resources/Countries_centroid.csv", encoding = "ISO-8859-1")
```

```python
un_migrant
```

```python
un_migrant.columns.tolist()
```

```python
un_migrant = un_migrant.drop(['Sort\norder','Notes','Code','Unnamed: 5','Total','Other North','Other South'], axis=1)
```

```python
un_migrant.rename(columns={'Major area, region, country or area of destination': "Country of Origin"}, inplace=True)
```

```python
un_migrant.set_index(["Year", "Country of Origin"])
```

```python
melt_un_migrant = pd.melt(un_migrant, id_vars=["Year", "Country of Origin"], 
                  var_name="Destination", value_name="numb_migrants")
```

```python
melt_un_migrant.head()
```

```python
melt_un_migrant[melt_un_migrant=='..'] = np.nan
melt_un_migrant = melt_un_migrant.dropna()
```

```python
melt_un_migrant['numb_migrants'] = pd.to_numeric(melt_un_migrant['numb_migrants'].str.replace(',',''))
```

```python
countries_df = countries_df.rename(columns={"name": "Country"})
countries_df.head()
```

```python
countries_list = list(countries_df["Country"])
countries_list
```

```python
melt_un_migrant

listOfAreas = countries_list
listOfAreas
```

```python
melt_un_migrant['Country of Origin'].isin(listOfAreas)

migrant_df = melt_un_migrant[melt_un_migrant['Country of Origin'].isin(listOfAreas)]
#First, we have list of areas
#then we get a list of true and false booleans based on whether the country of origin is in the lsit
#then we invert
#we pass this list of true and false to the original dataframe to filter and get the rows that do not have the list of areas in the column "country of origin"
```

```python
migrant_df = migrant_df.sort_values(["Year", "Country of Origin"])
migrant_df.head()
```

```python
migrant_df = migrant_df.reset_index(drop = True)
migrant_df.head()
```

```python
migrant_df = migrant_df.rename(columns={"Country of Origin" : "Origin"})
migrant_df.head()
```

```python
migrant_inflow = pd.DataFrame(migrant_df.groupby(["Year", "Origin"])["numb_migrants"].sum())
```

```python
migrant_inflow = migrant_inflow.rename(columns={"numb_migrants":"Inflow"})
new_migrant_inflow = migrant_inflow.reset_index()
new_migrant_inflow
list(new_migrant_inflow.columns)
#df.reset_index(level=['tick', 'obs'])
```

```python
countries_df_copy = countries_df.copy()
countries_df_copy = countries_df_copy.rename(columns={"Country": "Country"}).reset_index(drop=True)
countries_df_copy.head()
```

```python
#migrant_outflow_none = migrant_outflow[migrant_outflow['Origin'].isnull()]
#migrant_outflow_none
```

```python
migrant_inflow_m = pd.merge(new_migrant_inflow, countries_df_copy, left_on="Origin", right_on="Country", how="inner")
```

```python
migrant_inflow_m.head()
```

```python
migrant_outflow = pd.DataFrame(migrant_df.groupby(["Year", "Destination"])["numb_migrants"].sum())
```

```python
migrant_outflow = migrant_outflow.rename(columns={"numb_migrants":"Outflow"})
```

```python
new_migrant_outflow = migrant_outflow.reset_index()
```

```python
list(new_migrant_outflow.columns)
```

```python
migrant_outflow_m = pd.merge(new_migrant_outflow, countries_df_copy, left_on="Destination", right_on="Country", how="inner")
```

```python
list(migrant_outflow_m.columns)
```

```python
migrant_outflow_m.drop(columns=["Country"]).head()
```

```python
migrant_out_in = pd.merge(migrant_inflow_m, migrant_outflow_m, 
                          left_on=["Origin", "Year"], right_on=["Destination", "Year"], how="inner")
```

```python
migrant_out_in.head()
```

```python
migrant_out_in_1 = migrant_out_in[['Year','Country_x', 'Inflow', 'Outflow', 'latitude_x', 'longitude_x']]
```

```python
new_migrant_out_in = migrant_out_in_1.rename(columns={"latitude_x": "latitude", 
                                                     "longitude_x": "longitude", 
                                                     "Country_x": "country", 
                                                     "Year": "year", 
                                                     "Outflow": "outflow", 
                                                     "Inflow": "inflow"})
```

```python
new_migrant_out_in['net_migrants'] = new_migrant_out_in["inflow"] - new_migrant_out_in["outflow"]
```

```python
new_migrant_out_in = new_migrant_out_in.sort_values("year", ascending=True).set_index("year")
```

```python
migrant_org_df = migrant_df.rename(columns={"Year": "year", 
                                            "Origin": "origin", 
                                            "Destination": "destination"})
```

```python
migrant_org_lat_lng = pd.merge(migrant_org_df, countries_df_copy, left_on="origin", right_on="Country", how="inner")
```

```python
migrant_org_lat_lng = migrant_org_lat_lng.rename(columns={"latitude": "orig_lat", 
                                                           "longitude": "orig_lon"})
```

```python
migrant_org_lat_lng = migrant_org_lat_lng.drop("Country", axis=1)
```

```python
migrant_org_lat_lng.head()
```

```python
migrant_lat_lng = pd.merge(migrant_org_lat_lng, countries_df_copy, left_on="destination", right_on="Country", how="inner")
```

```python
migrant_lat_lng = migrant_lat_lng.rename(columns={"latitude": "dest_lat", 
                                                 "longitude": "dest_lon"})
```

```python
migrant_lat_lng = migrant_lat_lng.drop("Country", axis=1)
```

```python
migrant_lat_lng = migrant_lat_lng[["year", "destination", "origin", "numb_migrants", "dest_lat", "dest_lon", "orig_lat", "orig_lon"]]
#df = df[['mean', '0', '1', '2', '3']]
```

```python
migrant_lat_lng = migrant_lat_lng.sort_values("year", ascending=True).set_index("year")
```
