```python
import pandas as pd

df_1=pd.read_csv("Book1.csv")
df_2=pd.read_csv("Migrant_Inflow_Outflow_2.csv")

df_2.head(100)
df_3 = df_1.rename(columns={"Year":"year","Country Name":"country"})
df_4 = pd.merge(df_2,df_3,left_on = "country", right_on="country", how="inner")
df_2.head()
```

```python
df_5=df_4[['year_y','country', 'inflow', 'outflow', 'latitude', 'longitude',"net_migrants","Population"]]
df_rename= df_5.rename(columns={"year_y":"year"})
df_rename.head()
```

```python
df_6=pd.read_csv("GDP_per_capital.csv",encoding= "ISO-8859-1")
df_7 = df_6.melt(
    id_vars = ['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code'],
    value_vars = df_6.columns[4:], var_name = 'year').sort_values(['Country Name', 'year'])
df_7.head()
df_7.to_csv("test1")
```

```python

df_8 = df_7[["year","Country Name","value"]]
df_9 = df_8.set_index("year")
df_9 = df_9.reset_index()
df_9=df_9.rename(columns={"value":"GDP per Capita"})
df_9.head()
df_9.to_csv("test2.csv")
```

```python
df_10 = df_9.reset_index()
df_10.head()

year = [1990,1995,2000,2005,2010,2015,2017]
df_11 = df_10[pd.to_numeric(df_10["year"]).isin(year)]
```

```python
df_2['year']=df_2['year'].astype(int)
df_11['year']=df_11['year'].astype(int)
```

```python
df_2.columns, df_11.columns
```

```python
df_11 = df_11.rename(columns={"Country Name":"country"})
df_11.to_csv("test4.csv")
```

```python
df_12= pd.merge(df_2,df_11,on=["year","country"],how = "inner")
df_12.to_csv("test3.csv")
```

```python
df_13=df_12.sort_values(['country', 'year']).set_index("year")
df_13["GDP per Capita"].fillna("Data Unavailable",inplace = True)
```
